```python
# Selecting a subset of columns for violin plots for clarity
selected_columns = ['지가변동률', 'KB주택매매가격지수', '아파트매매가격지수', '한국은행 기준금리', '현금통화량', '주택담보대출-예금은행']


# Strip unnecessary spaces from column names
data.columns = data.columns.str.strip()

# Re-attempt to create violin plots with corrected column names
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 12))
axes = axes.flatten()

# Loop through the selected columns and plot violin plots
for i, column in enumerate(selected_columns):
    sns.violinplot(x=data[column], ax=axes[i])
    axes[i].set_title(column, fontsize=10)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Create a datetime index from the '기준년도' and '기준월'
data['날짜'] = pd.to_datetime(data['기준년도'].astype(str) + '-' + data['기준월'].astype(str))

# Plotting time series data for apartment sales and rental price indices
plt.figure(figsize=(14, 7))
plt.plot(data['날짜'], data['아파트매매가격지수'], label='아파트 매매가격지수', marker='o')
plt.plot(data['날짜'], data['아파트전세가격지수'], label='아파트 전세가격지수', marker='x')

plt.title('시간에 따른 아파트 매매가격지수와 전세가격지수의 변화')
plt.xlabel('날짜')
plt.ylabel('가격 지수')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = data.corr()

# Plotting the heatmap of the correlation matrix
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title('변수들 간의 상관 관계 히트맵')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
